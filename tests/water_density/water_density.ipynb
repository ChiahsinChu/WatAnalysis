```python
import MDAnalysis as mda
from MDAnalysis import transformations as trans

import matplotlib.pyplot as plt
```

```python
u = mda.Universe("../input_data/trajectory.xyz")
dim = [16.869, 16.869, 41.478, 90, 90, 120]
transform = trans.boxdimensions.set_dimensions(dim)
u.trajectory.add_transformations(transform)
```

# `Density1DAnalysis`

```python
from WatAnalysis.base import Density1DAnalysis
```

```python
density = Density1DAnalysis(u)
density.run()
```

```python
plt.plot(density.results["grids"], density.results["density"])
plt.ylim(bottom=0)
plt.xlim(density.results["grids"][0], density.results["grids"][-1])
plt.show()
```
