```python
import MDAnalysis as mda
from MDAnalysis import transformations as trans

import matplotlib.pyplot as plt
```

```python
u = mda.Universe("../input_data/pt_100.xyz")
dim = [11.246, 11.246, 35.94, 90, 90, 90]
transform = trans.boxdimensions.set_dimensions(dim)
u.trajectory.add_transformations(transform)
```

# `InterfaceWatOri`

```python
from WatAnalysis.base import InterfaceWatOri
```

```python
wat_ana = InterfaceWatOri(universe=u, slab_sel="name Pt", surf_natoms=16)
wat_ana.run()
```

```python
plt.plot(wat_ana.results["grids"], wat_ana.results["density"])
plt.ylim(bottom=0)
plt.xlim(wat_ana.results["grids"][0], wat_ana.results["grids"][-1])
plt.show()
```

```python
plt.plot(wat_ana.results["grids"], wat_ana.results["ori_dipole"])
plt.xlim(wat_ana.results["grids"][0], wat_ana.results["grids"][-1])
plt.show()
```
